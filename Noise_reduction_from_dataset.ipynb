## Below cell will replace the noisy data with clean data  

```python
for speech in tqdm(os.listdir(DATADIR)):
    rate, data = wavfile.read(os.path.join(DATADIR,speech))
    data = data / 32768
    try:  
        noise_reduced = nr.reduce_noise(audio_clip=data, noise_clip=data, prop_decrease=1.0, verbose=True)
        wavfile.write(os.path.join(DATADIR,speech),rate, noise_reduced)
    except:
        print("{} data is not formated perfectly".format(os.path.join(DATADIR,speech)))
```
